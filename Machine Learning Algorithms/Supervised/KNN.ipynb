```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
bc = datasets.load_breast_cancer()
x_train,x_test,y_train,y_test = train_test_split(bc.data,bc.target)
```

```python
def distance(x_train,y_train,x_test):
    m = len(x_train)
    n = len(x_test)
    list = []
    d = 0
    distance = []
    for i in range(n):
        list = []
        for j in range(m):
            #d = 0;
            d += ((x_test[i]- x_train[j])**2).sum()
            d = d**0.5
            list.append([d,j])
            list = sorted(list)
            list = list[0:5]
        distance.append(list)
    return distance

def predict(ds,y_train):
    y = []
    for i in range(len(ds)):
        yt = []
        count1 = 0
        count0 = 0
        for j in range(5):
            yt.append(y_train[ds[i][j][1]])
        for k in range(len(yt)):
            if(yt[k] == 0):
                count0 = count0 + 1
            else:
                count1 = count1 + 1
        if(count1 > count0):
            y.append(1)
        else:
            y.append(0)
    return y
    
def accuracy(y_test,y_pred):
    c1 = 0;
    m = len(y_test)
    for i in range(m):
        if(y_test[i] == y_pred[i]):
            c1 = c1 + 1
    return c1/len(y_test)*100
            
```

```python
ds = distance(x_train,y_train,x_test)
y_pred = predict(ds,y_train)
print(accuracy(y_test,y_pred))
plt.scatter(x_test[:,0],y_test)
plt.scatter(x_test,y_pred)
plt.show()
#print(ds[140][4])




#print(predict(ds,y_train))
#len(ds)
```
